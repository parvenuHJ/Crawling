### 멜론차트 TOP 100 수집하기

```python
# 라이브러리 import 할때 설명을 적는 습관을 들이자!
import requests as req  #  서버와의 요청 및 응답을 위한 라이브러리
from bs4 import BeautifulSoup as bs # html 문서에서 원하는 데이터를 추출하기 쉽게해주는 라이브러리
# 컴퓨터가 이해하기 쉬운 객체로 변환해주는 라이브러리
```

```python
url = 'https://www.melon.com/chart/index.htm'
req.get(url)
# 멜론페이지는 보안 문제 때문에 파이썬 코드로 접근하는 것을 막고 있다. -> <Response [406]>
# 그래서 우회접속을 해야함 ( 사용자인것처럼 속여야함 -> User Agent(사용자 대체자) )
```

```python
# 1. 400번대 에러 발생! -> 보안적으로 막고 있구나~
# 2. 접근하려는 페이지에서  F12(개발자모드)
# 3. NETWORK 탭으로 이동 
# 4. F5 눌러서 새로그침
# 5. 맨 위에 있는 document 타입의 html 문서 정보 열람
# 6. headers 탭의 맨 아래에 있는 User-Agent 값을 복사
# 7. 딕셔너리 형태로 파이썬 코드에 붙여넣기
# 8. 붙여넣은 결과를 url 옆에 headers값으로 지정!
```

```python
# 딕셔너리 형태 -->   {key : value ....}

# headers 의 약자로 h 로 변수명
h = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}
```

```python
# url 을 불러올때 사용자인것처럼 속여서 불러오겠다
res = req.get(url , headers = h)
res
```

```python
res.text
```

```python
# bs객체화 (beautiful soup 객체화) -> 파이썬이 알아듣게 객체화 => bs(parsing할 html문서, 'parsing할 방법')
soup = bs(res.text, 'lxml')
soup
```

### 데이터 수집
 - 노래 제목
 - 가수(아티스트)

```python
# 노래제목 수집하기
 
#   <div class =  ellipsis rank01 >    <span> <a> ~~~ </a> <span>    </div>
soup.select('div.ellipsis rank01>span>a')
```

```python
# 위 결과가 빈 리스트가 출력된다.  -> 이유 : ellipsis rank01 에서 '띄어쓰기' -> "자손선택자"로 인식해버림
# ellipsis 안의 rank01이라는 태그를 찾고있음 -> 존재X -> 빈 리스트 출력
# 선택자 이름 안에 공백은 자손선택자를 의미함. 그래서 rank01을 자손이라고 생각해버림

# 해결방법 : 공백을 온점으로 대체
soup.select('div.ellipsis.rank01>span>a')
# 의미 : div는 class를 2개 가지고 있다 ==> 1) ellipsis 2) rank01
# soup.select('div.ellipsis') 와 soup.select('div.rank01') 둘다 작동 가능
# 왜 이렇게 하냐? -> 클래스를 ellipsis 로 할때도 적용되고, rank01로 할때도 적용되게 하고 싶어서
```

```python
song = soup.select("div.ellipsis.rank01>span>a")
```

```python
len(song)
```

```python
# 가수이름(아티스트) 수집하기
# 공백은 . 으로 대체
singer = soup.select('div.ellipsis.rank02>a')
```

```python
len(singer)
```

- 가수 인원이 104 명이 나온 이유


```python
# 반복문 실습
# 1. 빈 공간 생성(빈 리스트 생성)
# 2. 컨텐츠 인덱스 번호 순서로 추출
# 3. 추출한 contents를 빈 공간에 한 개씩 차곡차곡 쌓기
```

```python
# 노래 이름 리스트 작성
songlist = [] # 1. 빈 공간 생성(빈 리스트 생성)
for i in range(len(song)) :  # 2. 컨텐츠 인덱스 번호 순서로 추출
    songlist .append(song[i].text) #3. ,  태그는 없이 컨텐츠만 넣어주도록 .text 사용
songlist
```


- div 태그로 하게되면 가수 이름이 두번씩 출력되니까
- 더 하위 태그인 span태그로

```python
# span 태그로
singer2 = soup.select('span.checkEllipsis')
```

```python
# 가수명 리스트 작성
singerlist = []
            # for i in range(len(song)) :  으로 해도 되지만 좀 다르게 해봄
for j in singer2:
    singerlist.append(j.text)
singerlist
# 페이지를 잘 살펴보는 습관을 들이자
```

- 위 결과에서 가수 이름이 2번이나 추출되는 이유
- div태그 안에 a태그와 span 태그가 같은 위치에 있음


```python
# 데이터프레임 형식으로 만들어보자
# 순위까지 추가해서!
ranklist = []  # 순위 리스트
for i in range(len(song)):
    ranklist.append(i+1)
ranklist
```

```python
# DataFrame 사용하기 위해 pandas 라이브러리 import
import pandas as pd
```

```python
dic = {'노래 제목' : songlist, '아티스트' : singerlist, '순위' : ranklist}
dic
```

```python
melon = pd.DataFrame(dic)
melon
```

```python
melon.set_index('순위', inplace = True) # inplace = True 를 하게되면 
                                                   # melon = melon.set_index('순위') 를 따로 안해줘도
                                                   # melon 을 출력하면 알아서 적용됨
```

```python
melon
```
