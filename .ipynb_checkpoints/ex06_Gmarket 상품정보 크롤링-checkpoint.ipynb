### Gmarket 상품정보 크롤링
 - 1위~10위까지 상품에 하나하나 접근(클릭메소드 이용)

```python
from selenium import webdriver as wb # 웹 브라우저를 통제하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys # 웹에게 값을 '입력'하기 위한 라이브러리
from selenium.webdriver.common.by import By # html 문서에서 태그와 선택자를 찾기 위한 라이브러리
from bs4 import BeautifulSoup as bs # 컴퓨터가 이해가 쉬운 객체로 변환해주는 라이브러리
import pandas as pd # 수집한 정보를 DataFrame화 하기 위한 라이브러리
import time # 코딩 중간에 쉬는 시간을 부여해주는 라이브러리
```

```python
# 1.
driver=wb.Chrome()
driver.get('https://www.gmarket.co.kr/n/best')
time.sleep(2)

# 2
namelist = []
pricelist = []
categorylist = []
rank = []
    
for i in range(0,10):
    driver.get('https://www.gmarket.co.kr/n/best') # 뒤로가기 쓰는 것보다 반복문 시작할떄 항상 페이지 로딩
    time.sleep(1) # 페이지 로딩 후 쉬는시간 
    btn_more = driver.find_elements(By.CLASS_NAME, 'itemname') # 상품명이 여러개니까 element's'
    btn_more[i].click() # 상품명이 여러개니까 버튼을 i번째 부터 클릭
    time.sleep(2)
    # bs 객체화
    soup = bs( driver.page_source , 'lxml')

    # 3
    name = soup.select('div.box__item-title>h1.itemtit')
    name[0].text
    price = soup.select('span.price_innerwrap>strong.price_real')
    price[0].text
    category = soup.select('div.location-navi>ul>li.on>a')
    category[0].text
    
    namelist.append(name[0].text)
    pricelist.append(price[0].text)
    categorylist.append(category[0].text)
    rank.append(i+1) # 1위~ 10위 순위 수집
    gmarket = pd.DataFrame(dic) # DataFrame 사용하기 위해 -> import pandas as pd

driver.close() # 크롬드라이버 종료
    
dic = {'이름' : namelist, '가격' : pricelist, '카테고리' : categorylist}

gmarket = pd.DataFrame(dic, index = rank) # index의 파라미터를 rank로 바꾸면 '0123' 말고 '123'가 인덱스가됨
gmarket
```

```python
# 선생님 코드

# 이 한 셀만 실행해도 크롤링이 자동으로 되도록 코딩
namelist = []
pricelist = []
catelist = []
rank = []

driver = wb.Chrome()
for i in range(1,11):
    driver.get('https://www.gmarket.co.kr/n/best') # 반복문이 실행될 때마다 페이지 refresh
    time.sleep(2)
    menu = driver.find_elements(By.CLASS_NAME, 'image__lazy.image__load')
    menu[i-1].click()  # 상품 1위부터 10위까지 클릭
    time.sleep(2)
    
    soup = bs(driver.page_source, 'lxml')  # 상품 상세 페이지 객체화
    m_name = soup.select_one('h1.itemtit')
    namelist.append(m_name.text)  # 상품명 수집
    m_price = soup.select_one('strong.price_real')
    pricelist.append(m_price.text)  # 상품가격 수집
    m_cate = soup.select_one('li.on>a')
    catelist.append(m_cate.text)  # 카테고리 수집
    rank.append(i)  # 1위 ~ 10위 순위 수집

driver.close() # 크롬드라이버 종료

dic = {'상품명' : namelist, '상품가격' : pricelist, '카테고리' : catelist}
data = pd.DataFrame(dic, index = rank)
data
```
