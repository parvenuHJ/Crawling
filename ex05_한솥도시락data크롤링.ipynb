```python
from selenium import webdriver as wb
# 웹브라우저를 제어하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys
# 웹에게 값을 '입력' 하기 위한 라이브러리 -> 컴퓨터용 키보드
from selenium.webdriver.common.by import By
# html 문서에서 태그와 선택자의 위치를 찾기 위한 라이브러리
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해하기 쉬운 객체로 변환해주는 라이브러리
```

## 한솥도시락 전체 메뉴 및 전체 가격 크롤링

```python
driver=wb.Chrome() # 크롬 켜라
driver.get('https://www.hsd.co.kr/menu/menu_list') # url 접속해라
```

```python
# bs 객체화 -> bs(parsing할 html문서, 'parsing 방법')
soup = bs( driver.page_source , 'lxml')
soup
```

```python
# 한솥도시락 메뉴, 가격 데이터 수집 실습
menu = soup.select('div.item-text>.h.fz_03')
menu
```

```python
menulist = []
for i in range(len(menu)) :
    menulist.append(menu[i].text)
menulist
```

```python
price = soup.select('div.item-price strong')
price
```

```python
pricelist = []
for i in range(len(price)) :
    pricelist.append(price[i].text + '원')
pricelist
```

```python
# DataFrame 사용하기 위해 pandas 라이브러리 import
import pandas as pd
```

```python
dic = {'메뉴이름' : menulist, '가격' : pricelist}
dic
```

```python
hansod = pd.DataFrame(dic)
hansod
```

#### 더보기 버튼을 눌러보자
 - 한솥 도시락 페이지는 더보기 버튼을 누르지 않으면 모든 메뉴가 보이지 않는다.
 - 더보기를 모두 눌러서 모든 메뉴를 수집해보자
 

```python
# 더보기 버튼의 위치를 찾아보자
btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
btn_more.click()
```

```python
# 더보기 버튼이 몇 번 중복되느닞 알 수 없다
# 몇 번 중복되는지 알 수 없기 때문에 15번 정도 눌러보자
```

```python
for i in range(15):
    btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
    btn_more.click()
    
print('안녕하세요')

# 15번 정도 눌러보고 안녕하세요 출력해라 -> 더보기가 15번이나 없기떄문에 오류남
```

### 예외처리 하는 방법
 - 실행할 부분(try)안에 넣어서 모두 실행하고, 만약 오류가 나면 예외처리(except)해주세요

```python
# 원래 코드 실행 시 에러가 발생하면 밑에 작성된 코드들은 중간에 멈춤
# 멈추지말고 계속해서 아래 코드를 실행할 수 있도록 예외처리 진행
try:
    # 에러가 나기 전까지 실행할 부분을 try문 안에 작성
    for i in range(15):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
except:
    # 에러가 났을 때 실행할 부분을 except문 안에 작성
    print('클릭 완료')
```

### time 라이브러리
 - 코딩 중간에 쉬는 시간을 부여해주는 라이브러리

```python
import time
# time.sleep(몇 초)
# 몇 초만큼 기다렸다가 다음 코드가 실행된다
# 사람이 처리하는 것처럼 해야 서버에 부하가 가지 않음
```

```python
try:
    # 에러가 나기 전까지 실행할 부분을 try문 안에 작성
    for i in range(15):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
        time.sleep(2)  # 2초 동안 쉬는시간 부여
except:
    # 에러가 났을 때 실행할 부분을 except문 안에 작성
    print('클릭 완료')
```
