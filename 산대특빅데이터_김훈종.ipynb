```python
# Question 3
# 1. G마켓 Best 1~20위까지 메뉴명, 가격, 카테고리를 수집하는 프로그램을 만드시오.
# 2. 수집한 데이터를 DataFrame으로 만드시오

# 라이브러리 불러오기
from selenium import webdriver as wb # 웹 브라우저를 통제하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys # 웹에게 값을 '입력'하기 위한 라이브러리
from selenium.webdriver.common.by import By # html 문서에서 태그와 선택자를 찾기 위한 라이브러리
from bs4 import BeautifulSoup as bs # 컴퓨터가 이해가 쉬운 객체로 변환해주는 라이브러리
import pandas as pd # 수집한 정보를 DataFrame화 하기 위한 라이브러리
import time # 코딩 중간에 쉬는 시간을 부여해주는 라이브러리

# 1.
driver=wb.Chrome()
driver.get('http://corners.gmarket.co.kr/Bestsellers')
time.sleep(2)

# 2
namelist = []
pricelist = []
categorylist = []
rank = []

for i in range(0,20):
    driver.get('http://corners.gmarket.co.kr/Bestsellers') 
    time.sleep(1) # 페이지 로딩 후 쉬는시간 
    btn_more = driver.find_elements(By.CLASS_NAME, 'itemname') # 상품명이 여러개니까 element's'
    btn_more[i].click() # 상품명이 여러개니까 버튼을 i번째 부터 클릭
    time.sleep(2)
    # bs 객체화
    soup = bs( driver.page_source , 'lxml')
    
    # 3
    name = soup.select('div.box__item-title>h1.itemtit')
    
    price = soup.select('span.price_innerwrap>strong.price_real')
    
    
    try :
        category = soup.select('div.location-navi>ul>li.on>a')
        categorylist.append(category[0].text)
    except :
        sub_category = soup.select('body > div.location-navi > ul > li:nth-child(3) > a')
        categorylist.append(sub_category[0].text)
        
    namelist.append(name[0].text)
    pricelist.append(price[0].text)
    rank.append(i+1) 

driver.close()

dic = {'이름' : namelist, '가격' : pricelist, '카테고리' : categorylist}
gmarket = pd.DataFrame(dic)

gmarket = pd.DataFrame(dic, index = rank)
gmarket
```

```python
# Question 4

import requests as req  #  서버와의 요청 및 응답을 위한 라이브러리
from bs4 import BeautifulSoup as bs # html 문서에서 원하는 데이터를 추출하기 쉽게해주는 라이브러리
import pandas as pd


url = 'https://www.melon.com/chart/index.htm'
req.get(url)

h = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'}

res = req.get(url , headers = h)

soup = bs(res.text, 'lxml')

songlist = [] 
singerlist = []
ranklist = []  
song = soup.select("div.ellipsis.rank01>span>a")
singer = soup.select('div.ellipsis.rank02>a')

for i in range(len(song)) :  # 2. 컨텐츠 인덱스 번호 순서로 추출
    songlist .append(song[i].text) #3. ,  태그는 없이 컨텐츠만 넣어주도록 .text 사용
    singerlist.append(singer[i].text)
    ranklist.append(i+1)

dic = {'노래 제목' : songlist, '아티스트' : singerlist, '순위' : ranklist}
melon = pd.DataFrame(dic)
melon.set_index('순위', inplace = True)
melon.to_csv('멜론차트.txt', encoding = 'euc-kr')
melon
```

```python
# Question 4
from selenium import webdriver as wb
# webdriver : 웹브라우저를 제어하기 위한 라이브러리
from selenium.webdriver.common.keys import Keys
# Keys : 웹에게 값을 '입력' 하기 위한 라이브러리 -> 컴퓨터용 키보드
from selenium.webdriver.common.by import By
# By : html 문서에서 태그와 선택자의 위치를 찾기 위한 라이브러리, 컨텐츠가 있는 위치를 찾아줌
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해하기 쉬운 객체로 변환해주는 라이브러리
import pandas as pd

driver=wb.Chrome() 
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(2)

try:
    # 에러가 나기 전까지 실행할 부분을 try문 안에 작성
    for i in range(15):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
        time.sleep(2)  # 2초 동안 쉬는시간 부여
except:
    # 에러가 났을 때 실행할 부분을 except문 안에 작성
    print('클릭 완료')
    
soup = bs( driver.page_source , 'lxml')
menu = soup.select('div.item-text>.h.fz_03')
price = soup.select('div.item-price strong')


menulist = []
pricelist = []
for i in range(len(menu)) :
    # 전처리 (['8월~ ' 문자 없애기])
    temp = menu[i].text
    if "8월" in temp:
        temp = temp[9:]
    menulist.append(temp)
    pricelist.append(price[i].text + '원')
    
dic = {'메뉴이름' : menulist, '가격' : pricelist}
hansod = pd.DataFrame(dic)
hansod.to_csv('한솥.txt', encoding = 'euc-kr')
hansod
```
