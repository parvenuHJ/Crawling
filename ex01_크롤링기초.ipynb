### requests 라이브러리 사용법

```python
# 라이브러리 불러오기
import requests as req
# 서버에게 데이터를 요청 및 응답받기 위한 라이브러리
```

```python
res = req.get("https://www.naver.com/")
# <Response [200번대]> : 성공적으로 요청 및 응답이 이루어졌다는 의미
# <Response [400번대]> : 클라이언트(요청) 부분에서 오류가 발생했다는 의미
# <Response [500번대]> : 서버(응답) 부분에서 오류가 발생했다는 의미
```

```python
# html 문서형태로 가져와진다
res.text
```

## BeautifulSoup 라이브러리
### : html 문서에서 원하는 데이터를 추출하기 쉽게 해주는 라이브러리
### : 컴퓨터가 이해할 수 있는 파이썬 객체 형태로 변환해주는 라이브러리

```python
# BeautifulSoup 라이브러리가 너무 광범위해서 from 키워드 사용해줌
from bs4 import BeautifulSoup as bs
```

```python
res.text # text로 변환된 html 문서
```

```python
# bs(parsing할 html문서, 'parsing할 방법') → 객체화
# lxml : 빠르고 관대하다는 특징이 있음
# parsing 을 해서 파이썬이 이해할 수 있게 만들어줌
soup = bs(res.text , 'lxml')
soup
```

##### 네이버 페이지에서 정보를 가져와보자
- 뉴스, 인플루언서, 통합 등과 같은 글자를 수집해보자

```python
# a태그를 가져오는 방법
soup.select('a')
# 페이지에 있는 모든 a태그가 가져와짐
```

```python
# a태그 중에 클래스가 tab인 애들만 데려와보자
naver_tab = soup.select('a.tab') # -> [ , , ] : list형태
naver_tab
```

```python
# 여기 중에 '통합'이라는 contents를 가져와보자 1
naver_tab[0] 
```

```python
# 여기 중에 '통합'이라는 contents를 가져와보자 2
# select_one : 여러가지 요소 중에서 첫번째 요소만 "단수형"으로 가져와준다 (단수형 -> listX)
# select : 여러가지 요소를 "복수형"으로 가져와준다 (복수형 -> list O)
soup.select_one('a.tab')
```

```python
# <a aria-sel ~ role="tab">통합</a> 중에서     "문자" 만 가져오겠다 
naver_tab[0].text
```

### 데이터 수집 실습
- 네이버에 '날씨' 라고 검색한 후 온도 데이터를 수집 

```python
# 결과값 : 현재 온도 33.1º  로 출력해보기
```

```python
# 1. 페이지 정보 받아오기
res_weth = req.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%82%A0%EC%94%A8')
```

```python
res_weth
```

```python
res_weth.text
```

```python
# 2. 객체화 진행(컴퓨터가 이해하기 쉬운 파이썬 객체로 변환)
soup_weth = bs(res_weth.text, 'lxml')
soup_weth
```

```python
#  3. 태그 선택하기  ( 중요 !!!!!!!!!!!!!!!!!!!! )

#  <strong><span class="blind">현재 온도</span>34.1<span class="celsius">°</span></strong>
# 여기에 있는 span 태그를 모두 가져오고 싶다 -> <strong> 태그를 가져와야겠다 
# <strong> 태그에 "class 선택자가 없어서 이것만 추려서 가져올 수 없음" !!!
# -> 그럼 바로 한 단계 윗단계인 "<div class="temperature_text"> ~ </div>" 를 사용!
soup_weth.select('div.temperature_text')
```

```python
# 4. 컨텐츠 추출하기
soup_weth.select_one('div.temperature_text').text
print(soup_weth.select_one('div.temperature_text').text)
```
